## 1.1 Tratamiento de Datos Faltantes

---
Limpieza de datos e imputación:

1. Se imputa por MINIMUM_PAYMENTS consdierando la agrupación por LÍMITE DE CRÉDITO. Es decir, agrupamos a los clientes por LÍMITE DE CRÉDITO y utilizamos la media de cada grupo para completar los valores faltantes.


```python
from sklearn.impute import KNNImputer
```

```python
df["MINIMUM_PAYMENTS"] = df.groupby(pd.cut(df['CREDIT_LIMIT'],4))['MINIMUM_PAYMENTS'].transform(lambda x: x.fillna(x.mean()))
```

---
2. Se realiza nuevamente una imputación (secundaria con KNN), para manejar los valores nulos restantes en 'MINIMUM_PAYMENTS'.


**Como Funciona el KNNImputer**

El algoritmo identifica las filas que tienen valores faltantes.

Para cada valor faltante, el algoritmo busca los K vecinos mas cercanos. Los vecinos se definen en funcion de la distancia. Las instancias que tienen caracteristicas familiares se consideran vecinos cercanos.

Unz vez identificados los K vecinos mas cercanos, el valor faltante se imputa con el promedio de los valores de las caracteristicas de esos vecinos.

```python
if df['MINIMUM_PAYMENTS'].isnull().sum() > 0:
    knn_imputer_secondary = KNNImputer(n_neighbors=5)
    imputed_data = knn_imputer_secondary.fit_transform(df[['MINIMUM_PAYMENTS', 'CREDIT_LIMIT', 'BALANCE', 'PAYMENTS']])
    df['MINIMUM_PAYMENTS'] = pd.DataFrame(imputed_data, columns=['MINIMUM_PAYMENTS', 'CREDIT_LIMIT', 'BALANCE', 'PAYMENTS'])['MINIMUM_PAYMENTS']
```

```python
df.head()
```

---
3. Ahora, se realiza una imputación de 'CREDIT_LIMIT' según la agrupación de la variable 'PURCHASES'. Es decir, agrupamos a los clientes por 'CREDIT_LIMIT' y utilizamos la media de cada grupo para completar, de igual manera, los valores faltantes.

```python
## Imputar CREDIT_LIMIT según la agrupación por PURCHASES. Agrupar a los clientes por LÍMITE DE CRÉDITO y utilizar la mediana de cada grupo para completar los valores faltantes
df['CREDIT_LIMIT'] = df.groupby(
    pd.qcut(df['PURCHASES'], 4)
)['CREDIT_LIMIT'].transform(lambda x: x.fillna(x.mean()))
```

```python
## Manejar los valores nulos restantes en CREDIT_LIMIT (imputación secundaria con KNN)
if df['CREDIT_LIMIT'].isnull().sum() > 0:
    knn_imputer_credit = KNNImputer(n_neighbors=5)
    imputed_credit_data = knn_imputer_credit.fit_transform(df[['CREDIT_LIMIT', 'BALANCE', 'PURCHASES', 'PAYMENTS']])
    df['CREDIT_LIMIT'] = pd.DataFrame(imputed_credit_data, columns=['CREDIT_LIMIT', 'BALANCE', 'PURCHASES', 'PAYMENTS'])['CREDIT_LIMIT']
```

## 3. Detección y Eliminación de Valores atípicos
---
 Identificamos posibles valores atípicos en el conjunto de datos utilizando el método estadístico Z score y Rango intercuartil, para seleccionar el menos invasivo.

```python
df_filtered.info()
```

```python
columns = df_filtered.select_dtypes(include=[np.number]).columns
filtered_data_zscore = df_filtered.copy()
filtered_data_iqr = df_filtered.copy()
outliers_per_column_zscore = {}
outliers_per_column_iqr = {}
```

##Graficamos para entender mejor los datos.

```python
plt.figure(figsize=(15, 15))
for i, col in enumerate(df_filtered.select_dtypes(include=['number']).columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df_filtered[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.show()
```

```python
# Create the boxplots
plt.figure(figsize=(15, 10))  # Adjust figure size for better visualization
for i, col in enumerate(df_filtered.select_dtypes(include=['number']).columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df_filtered[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```

##Z-Score

Al utilizar Z-score, con tres desviaciones estándar obtenemos un número de outliers de:

```python
# Z-score
for column in columns:
    z_scores = np.abs((df_filtered[column] - df_filtered[column].mean()) / df_filtered[column].std())
    outliers = z_scores > 3
    outliers_per_column_zscore[column] = outliers.sum()
    filtered_data_zscore = filtered_data_zscore[~outliers]
```

```python
filtered_data_zscore.shape
```

```python
## Graficos sin datos con una z-score >3
plt.figure(figsize=(15, 15))
for i, col in enumerate(filtered_data_zscore.select_dtypes(include=np.number).columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(filtered_data_zscore[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.show()
```

Rango IQ

```python
for column in columns:
    Q1 = df_filtered[column].quantile(0.25)
    Q3 = df_filtered[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = (df_filtered[column] < (Q1 - 1.5 * IQR)) | (df_filtered[column] > (Q3 + 1.5 * IQR))
    outliers_per_column_iqr[column] = outliers.sum()
    filtered_data_iqr = filtered_data_iqr[~outliers]
```

```python
filtered_data_iqr.shape
```

## 2.1 Aplicar Clustering

---
Se aplica el método del 'Codo' para determinar el número apropiado de cluster para nuestra base de datos.

```python
normalized_columns_logs
```

```python
from sklearn.cluster import KMeans
```

```python
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_data_log[normalized_columns_logs])
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.grid()
plt.show()
```

Se determinan 4 cluster como el número apropiado, de esta manera el siguiente código lleva este número base.
---

```python
k_optimal = 4
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
normalized_data_log['Cluster'] = kmeans.fit_predict(normalized_data_log[normalized_columns_logs])
```
