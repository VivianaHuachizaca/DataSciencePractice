1. **Análisis de dataset simplificado**:
   Dada una lista de diccionarios que representa un dataset (por ejemplo, estudiantes con “Nombre”, “Edad”, y “Nota Final”), filtra y muestra los estudiantes con una nota mayor a 8.

```python
estudiantes = [
    {"Nombre": "Ana", "Edad": 18, "Nota Final": 9.5},
    {"Nombre": "Luis", "Edad": 20, "Nota Final": 8.7},
    {"Nombre": "María", "Edad": 19, "Nota Final": 8.9},
    {"Nombre": "Pedro", "Edad": 22, "Nota Final": 6.9},
    {"Nombre": "Sofía", "Edad": 24, "Nota Final": 8.2},
]
estudiantes_con_8 = list(filter(lambda estudiantes: estudiantes["Nota Final"] > 8, estudiantes))

for estudiante in estudiantes_con_8:
    print(("Los estudiantes con notas mayor a 8 son:"))
    print(f"Nombre: {estudiante['Nombre']}, Edad: {estudiante['Edad']}, Nota Final: {estudiante['Nota Final']}")
```

2. **Limpieza de datos duplicados**:
   Diseña un programa que elimine los valores duplicados de una lista de datos. Usa un conjunto para esto y transforma el resultado nuevamente en lista.

```python
def no_duplicados(lista):
    lista_sin_duplicados = []
    for elemento in lista:
        if elemento not in lista_sin_duplicados:
            lista_sin_duplicados.append(elemento)
    return lista_sin_duplicados


datos = [1,1,1,1, 2, 2, 3, 4, 4, 5, 6, 6, 6, 6, 7, 8, 8, 9, 9]
resultado = no_duplicados(list(set(datos)))
print("Lista sin duplicados:", resultado)
```

3. **Frecuencia de palabras**:
   Dado un texto de varias líneas, crea un diccionario que cuente cuántas veces aparece cada palabra. Ignora mayúsculas/minúsculas y los signos de puntuación.

```python
def contar_palabras(texto):
    texto = texto.lower()
    palabras = texto.split()
    frecuencias = {}
    for palabra in palabras:
        if palabra in frecuencias:
            frecuencias[palabra] += 1
        else:
            frecuencias[palabra] = 1

    return frecuencias

texto = """Las clases virtuales se han convertido en un desafío por los cortes de energía constantes en todas las ciudades del país."""

conteo = contar_palabras(texto)

for palabra, frecuencia in conteo.items():
    print(f"{palabra}: {frecuencia}")
```

4. **Agrupación de datos**:
   Escribe un programa que agrupe una lista de edades en rangos (por ejemplo, 0-18, 19-35, 36-50, 51+). Usa un diccionario para almacenar las edades agrupadas por rango.

```python
def grupo_edades(edades):
    grupos = {
        "0-18": [],
        "19-30": [],
        "31-50": [],
        "51+": []
    }
    for edad in edades:
        if 0 <= edad <= 18:
            grupos["0-18"].append(edad)
        elif 19 <= edad <= 30:
            grupos["19-30"].append(edad)
        elif 31 <= edad <= 50:
            grupos["31-50"].append(edad)
        elif edad >= 51:
            grupos["51+"].append(edad)

    return grupos

edades = [2, 5, 8, 12, 17, 22, 31, 34, 45, 51, 67, 80, 18, 36, 19, 55, 24, 89]
grupos = grupo_edades(edades)

# Mostrar el resultado
for rango, grupo in grupos.items():
    print(f"{rango}: {grupo}")
```

5. **Generación de gráficos**:
   Usando listas de datos (ingresos y gastos mensuales), escribe funciones para calcular la utilidad neta y generar una tabla de resultados. Usa librerías como matplotlib para visualizar las tendencias (opcional si están familiarizados con gráficos).

```python
import matplotlib.pyplot as plt
```

```python
def utilidad_neta(ingresos, gastos):
    if len(ingresos) != len(gastos):
        raise ValueError("Las listas de ingresos y gastos deben tener la misma longitud.")
    return [ing - gas for ing, gas in zip(ingresos, gastos)]

def resultados(ingresos, gastos, utilidades):
    print("Mes\tIngresos\tGastos\t\tUtilidad Neta")
    print("-" * 40)
    for i, (ing, gas, util) in enumerate(zip(ingresos, gastos, utilidades), 1):
        print(f"{i}\t{ing:.2f}\t\t{gas:.2f}\t\t{util:.2f}")

def grafico(ingresos, gastos, utilidades):
    meses = list(range(1, len(ingresos) + 1))
    plt.figure(figsize=(10, 6))
    plt.plot(meses, ingresos, label="Ingresos", marker='o')
    plt.plot(meses, gastos, label="Gastos", marker='o')
    plt.plot(meses, utilidades, label="Utilidad Neta", marker='o')
    plt.title("Tendencias Financieras Mensuales")
    plt.xlabel("Mes")
    plt.ylabel("Monto ($)")
    plt.legend()
    plt.grid()
    plt.show()

ingresos = [5000, 8500, 2850, 4000, 4200, 6100, 4300, 4100, 4500, 4600]
gastos = [3000, 4000, 2200, 2300, 2500, 5400, 2600, 2700, 8540, 1500]
utilidades = utilidad_neta(ingresos, gastos)

# Generar tabla de resultados
resultados(ingresos, gastos, utilidades)

# Graficar tendencias
grafico(ingresos, gastos, utilidades)
```
